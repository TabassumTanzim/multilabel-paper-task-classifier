```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
webdriver_path = r"/Users/tabassumtanzim/Downloads/chromedriver-mac-arm64/chromedriver"
options = webdriver.ChromeOptions()
options.add_argument(f"webdriver.chrome.driver={webdriver_path}")
```

```python
df=pd.read_csv("21k.csv")
df
```

```python
#converting to list
urls = df.url.to_list()

title=df.title.to_list()

paper_details=[]
```

```python

# Open the WebDriver
driver = webdriver.Chrome(options=options)

for idx,url in enumerate(urls[14553:15000]):
    driver.get(url)
    time.sleep(2)

    try:
        # Title
        section = driver.find_element(By.CLASS_NAME, "paper-title")
        heading = section.find_element(By.TAG_NAME, "h1")
        title = heading.text
#         print("TITLE", title)

        # Abstract
        paper = driver.find_element(By.CLASS_NAME, "paper-abstract")
        col = paper.find_elements(By.CLASS_NAME, "col-md-12")
        abstract = paper.find_element(By.TAG_NAME, "p")
#         print("ABSTRACT: ", abstract.text)

        # Methods
        task = driver.find_element(By.CLASS_NAME, "paper-tasks")
        col_elements = task.find_elements(By.CLASS_NAME, "badge-primary")

        # Initialize an empty list to store the results
        result_list = []

        # Iterate through each element to extract the information from the span tag
        for col_element in col_elements:
            # Find the "span" tag within the current element
            span_element = col_element.find_element(By.TAG_NAME, "span")

            # Get the text content of the "span" tag
            tasks = span_element.text

            # Append the text content to the result list
            result_list.append(tasks)

        # Print the result list
#         print("TASKS", result_list)

        # Append paper details to the list
        paper_details.append({
            "title": title,
            "url": url,
            "abstract": abstract.text,
            "tasks": result_list
        })

    except Exception as e:
        # Handle the exception (print or log the error, or skip this iteration)
        print(f"Error processing URL {url}: {e}")
    print(idx)
# Close the WebDriver
driver.quit()
```
